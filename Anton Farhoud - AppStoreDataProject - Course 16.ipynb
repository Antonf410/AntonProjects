```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('C:\\Users\\ASUS\\Python\\Datasets\\appstore_games.csv')
df
```

```python
# The data I will be working on is a database of game applications in App Store (Apple).
# The database contains 17,007 rows and 18 different columns.
df.shape
```

```python
df.columns
```

```python
# Deleting columns because they don't have much meaning.
df.drop(['URL','Subtitle','Icon URL','Description','In-app Purchases'],axis=1,inplace=True)
df.isnull().sum()
```

```python
# Rename columns:
# Price to 'Price $'
# Size to 'Size Byte'
df.rename(columns={'Price':'Price ($)','Size':'Size (Byte)'},inplace=True)
df
```

```python
# Why I keep the remaining columns?
# ID - it's ID number for each applications.
# Name - The name of the app. It helps you understand which app the data in each row refers to.
# Average User Rating - This column is important if we're trying to understand which apps are highly rated.
# User Rating Count - The number of ratings can show an app with many ratings is generally considered more reliable and successful.
# Price ($) - The price of the app is particularly important for market and pricing analyses.
#         here can explore the relationship between price and popularity or quality.
# Developer - Name of the developer.
# Age Rating - The "Age Rating" column indicates the recommended minimum age for users of the app or game.
# Languages - how many languages the apps have.
# Size (Byte) - to know if have a space in storage to download the application.
# Primary Genre - The app's primary genre helps categorize the apps into specific types of games or areas, which can be important for analysis by category.
# Genres - This field allows to see what additional categories are associated with the app, which can be important for niche or genre-specific analyses.
# Original Release Date - The original release date allows to analyze the data over time, for example,
#                         to see which older apps are still successful or to observe trends in game development over time.
# Current Version Release Date - This is the date when the latest update.
```

```python
df.info()
```

```python
# here I change the column "Original Release Date" & "Current Version Release Date" to datetime format
df['Original Release Date']=pd.to_datetime(df['Original Release Date'],dayfirst=True)
df['Current Version Release Date']=pd.to_datetime(df['Current Version Release Date'],dayfirst=True)
df.info()
```

```python
# Add Original Release Year column
df['Original Release Year']=df['Original Release Date'].dt.year
```

```python
# Add Update Year column
df['Update Year']=df['Current Version Release Date'].dt.year.round(1)
```

```python
# Showing the difference in year between the 'Original Release Date' and the date of the last update of the application.
diffyears=df['Current Version Release Date']-df['Original Release Date']
df["Different Years"]=(diffyears.dt.days/365).round(1)
df[['Name','Original Release Year','Update Year','Different Years']]
```

```python
# Show count application for each genre
df['Primary Genre'].value_counts()
```

```python
# Show count application for each genre in countplot
sb.countplot(x='Primary Genre',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# The graph above shows the clear dominance of the gaming genre in the app market,
# with a huge amount of apps in this area compared to the other genres.
# The other genres contain far fewer apps, indicating a strong focus on games by developers.
# which indicates the high popularity and involvement of the users of gaming apps
```

```python
# Calculates the 'Average User Rating' for each genre
df.groupby('Primary Genre')['Average User Rating'].mean().reset_index()
```

```python
# Calculates the 'Average User Rating' for each genre, showing in barplot
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
avg_genre=df.groupby('Primary Genre')['Average User Rating'].mean().reset_index()
cmap=LinearSegmentedColormap.from_list("red_green", ["red", "yellow", "green"])
norm=plt.Normalize(avg_genre['Average User Rating'].min(), avg_genre['Average User Rating'].max())
colors=cmap(norm(avg_genre['Average User Rating']))
colorsls=colors.tolist()
plt.figure(figsize=(10,5))
plt.title('Average User Rating by Genre',fontsize=20,color='red')
sb.barplot(x='Primary Genre',y='Average User Rating',data=avg_genre,hue='Primary Genre',palette=colorsls)
plt.xticks(rotation=90)
plt.show()
```

```python
# The graph above shows the distribution of the average user ratings by genre
# and shows that there are genres that receive higher popularity from users,
# such as food and drink or music, and other genres that receive lower ratings, such as business and medicine.
# This information can be used to understand which genres are considered higher quality and in demand by users.
```

```python
# Showing trend of Average User Rating Over Time, line
avg_year=df.groupby('Original Release Year')['Average User Rating'].mean().reset_index()
plt.figure(figsize=(10,5))
plt.title('Trend of Average User Rating Over Time',fontsize=20,color='red')
plt.xticks([2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.xlabel('Year')
sb.lineplot(x='Original Release Year',y='Average User Rating',data=avg_year,marker='o', markersize=8)
plt.show()
```

```python
# The graph above shows a strong improvement between 2008 and 2012,
# which may indicate that developers are improving their products,
# or that apps have become better, or that user expectations and experiences are changing and influencing how they rate apps.
# After 2012 there is a certain stabilization in the ratings, with a slight increase every year.
```

```python
# showing paid applications
paidapps=df[df['Price ($)']>0]
paidapps[['Name','Price ($)']]
```

```python
# showing MAX Price of App
paidapps.groupby('Name')['Price ($)'].max().reset_index().sort_values(by='Price ($)',ascending=False)
```

```python
# showing free applications
freeapps=df[df['Price ($)']==0]
freeapps[['Name','Price ($)']]
```

```python
# showing quantity of applications by price
plt.figure(figsize=(10,5))
plt.title('Quantity of applications by price',fontsize=20,color='red')
sb.countplot(x='Price ($)',data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# The graph above shows a clear trend where free and cheap apps dominate the market.
# High-priced apps are rare, likely aimed at special purposes or offering advanced features that justify the cost.
```

```python
# create data from with to columns 'Average User Rating' for free apps and paid apps
data={'App Type':['Free Apps','Paid Apps'],
      'Average Rating':[freeapps['Average User Rating'].mean(),paidapps['Average User Rating'].mean()]}
data=pd.DataFrame(data)
data
```

```python
# showing different rate between paid application and free application in pie
plt.figure(figsize=(10,5))
plt.title('Diffirent Rate Between Paid Apps and Free Apps',fontsize=18,color='red')
plt.pie(data['Average Rating'],labels=data['App Type'],autopct='%1.1f%%')
plt.show()
```

```python
# The pie chart above shows that there is an almost equal amount of free and paid apps in the market,
# suggesting that users are exposed to both types of business models similarly.
# This can be especially important in market analysis to understand which types of apps offer more opportunities for developers.
```

```python
# showing how count apps per each developer
DevApps=df['Developer'].value_counts().reset_index()
DevApps.columns=['Developer','Num. Apps']
DevApps
```

```python
# age rate count
age_rate=df['Age Rating'].value_counts().reset_index()
age_rate.columns=['Age Rating','Num. Apps']
age_rate
```

```python
age_genre=df.groupby(['Primary Genre','Age Rating']).size().unstack(fill_value=0)
sb.heatmap(age_genre,cmap='Blues',annot=True,fmt='d',cbar=False,linewidths=0.4, linecolor='black')
plt.title('Number of Apps by Genre and Age Rating', fontsize=18,color='red')
plt.xlabel('Age Rating')
plt.ylabel('Genre')
plt.show()
```

```python
# We can see above,
# The "games" genre is distinctly dominant in terms of the number of apps in all age ratings,
# especially in the '4+' and '12+' categories.
# The vast majority of genres focus on apps rated '4+',
# indicating that these apps are intended for a wider audience, including children.
# Other genres such as "education" and "entertainment" contain a significant number of apps
# but are less common compared to the gaming genre.
```

```python
app_cnt_year=df.groupby('Original Release Year').size().reset_index(name='Num. Apps')
plt.figure(figsize=(10,5))
plt.title('Number of Apps by Year',fontsize=20,color='red')
sb.lineplot(x='Original Release Year',y='Num. Apps',data=app_cnt_year,marker='o', markersize=8)
plt.xticks([2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.show()
```

```python
# We can see above a clear upward trend in the number of apps published every year from 2008 until about 2016.
# This indicates a period of rapid increase in the app market during these years.
# The year 2016 marks the peak in the number of published apps,
# with the highest number of apps being launched this year.
# This may be due to the maturing of the market,
# with more developers creating apps to meet the growing demand.
# After 2016 the number of apps is on the decrease,
# developers may have shifted their focus from quantity to quality,
# preferring to improve and update existing apps instead of adding a large amount of new apps to the market.
```

```python
app_update_count=df.groupby('Update Year').size().reset_index(name='Num. Apps of Update')
plt.figure(figsize=(10,5))
plt.title('Number of Updatinf Apps by Year',fontsize=20,color='red')
sb.lineplot(x='Update Year',y='Num. Apps of Update',data=app_update_count,marker='o', markersize=8)
plt.xticks([2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])
plt.show()
```

```python
# We can see above that after 2015 there is an increase in updating the applications.
# As we saw in the previous graph the decrease in the number of applications after 2016,
# at the same time you see in the graph above the number of updates of the increasing applications.
# This may be due to the increasing importance of maintaining the relevance and quality of applications.
```
